## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

## Pull the Data

```python
income_df = pd.read_csv('2023StateIncome.csv')
income_df = income_df.dropna()
```

```python
homeValue_df = pd.read_csv('Data/yearlyStateValue.csv')
homeValue_df = homeValue_df.dropna()
homeValue_df['state'] = homeValue_df['RegionName']
```

## Merge DataFrames on State

```python
combined_df = pd.merge(income_df, homeValue_df, on='state', how='inner')
```

```python
combined_df = combined_df[['state','2023','HouseholdIncome']]
clean_df['HomeValue'] = combined_df['2023']
clean_df
```

```python
clean_df['affordability'] = clean_df['HomeValue'] / clean_df['HouseholdIncome']
clean_df.drop(columns='2023')
```
